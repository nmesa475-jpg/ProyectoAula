```python
import re

def Menu():
    print('Seleccione una de las opciones:')
    print('1. Mostrar canciones')
    print('2. Agregar canción')
    print('3. Borrar canción')
    print('4. Buscar canciones')
    print('5. Mostrar tiempo total')
    print('')
    print('0. Salir')
    print('')

def segundos(duracion):
    """Convierte 'mm:ss' a segundos totales"""
    try:
        minutos, segundos = map(int, duracion.split(':'))
        return minutos * 60 + segundos
    except:
        return None

def minutosasegundos(segundos):
    """Convierte segundos a formato 'mm:ss'"""
    minutos = segundos // 60
    segundos = segundos % 60
    return f"{minutos:02d}:{segundos:02d}"

def mostrarcanciones(biblio):
    print('\nLista de canciones:')
    for i, cancion in enumerate(biblio, 1):
        print(f"{i}. {cancion['Nombre']} - {cancion['Artista']} ({cancion['Año']}) [{cancion['Duración']}]")
    print()

def agregarcancion(biblio):
    print('\nAgregar nueva canción:')
    nombre = input('Nombre: ')
    artista = input('Artista: ')
    año = input('Año: ')
    duracion = input('Duración (mm:ss): ')

    if segundos(duracion) is None:
        print('Formato de duración inválido. Use mm:ss\n')
        return

    genero = input('Género: ')
    idioma = input('Idioma: ')

    nueva_cancion = {
        'Nombre': nombre,
        'Artista': artista,
        'Año': año,
        'Duración': duracion,
        'Género': genero,
        'Idioma': idioma
    }

    biblio.append(nueva_cancion)
    print(f'Canción "{nombre}" agregada correctamente\n')

def buscarcanciones(biblio):
    print('\nBuscar canciones:')
    termino = input('Buscar por nombre, artista o género: ').lower()

    resultados = []
    for cancion in biblio:
        if (termino in cancion['Nombre'].lower() or
            termino in cancion['Artista'].lower() or
            termino in cancion['Género'].lower()):
            resultados.append(cancion)

    if not resultados:
        print('No se encontraron resultados\n')
        return

    print(f'\nSe encontraron {len(resultados)} canción(es):')
    for cancion in resultados:
        print(f"- {cancion['Nombre']} - {cancion['Artista']} ({cancion['Año']})")
    print()

def borrarcancion(biblio):
    if not biblio:
        print('No hay canciones para borrar\n')
        return

    mostrarcanciones(biblio)
    try:
        seleccion = int(input('Número de la canción a borrar: '))
        if 1 <= seleccion <= len(biblio):
            cancion_eliminada = biblio.pop(seleccion - 1)
            print(f'Canción "{cancion_eliminada["Nombre"]}" eliminada\n')
        else:
            print('Número inválido\n')
    except:
        print('Entrada inválida\n')

def mostrartiempototal(biblio):
    total_segundos = 0
    for cancion in biblio:
        segundos_cancion = segundos(cancion['Duración'])
        if segundos_cancion is not None:
            total_segundos += segundos_cancion

    print(f'\nTiempo total: {minutosasegundos(total_segundos)}\n')

# Programa principal
biblioteca = [
    {"Nombre": "Imagine",        "Artista": "John Lennon",          "Año": "1971", "Duración": "03:03", "Género": "Pop",       "Idioma": "Inglés"},
    {"Nombre": "Bésame Mucho",   "Artista": "Consuelo Velázquez",   "Año": "1940", "Duración": "02:50", "Género": "Bolero",    "Idioma": "Español"},
    {"Nombre": "Shape of You",   "Artista": "Ed Sheeran",           "Año": "2017", "Duración": "03:53", "Género": "Pop",       "Idioma": "Inglés"},
    {"Nombre": "La Flaca",       "Artista": "Jarabe de Palo",       "Año": "1996", "Duración": "04:30", "Género": "Rock/Pop",  "Idioma": "Español"},
    {"Nombre": "Clocks",         "Artista": "Coldplay",             "Año": "2002", "Duración": "05:07", "Género": "Rock",      "Idioma": "Inglés"},
]

while True:
    Menu()
    opcion = input('Digita la opción a escoger: ')

    match opcion:
        case '1':
            mostrarcanciones(biblioteca)
        case '2':
            agregarcancion(biblioteca)
        case '3':
            borrarcancion(biblioteca)
        case '4':
            buscarcanciones(biblioteca)
        case '5':
            mostrartiempototal(biblioteca)
        case '0':
            print('¡Hasta luego!')
            break
        case _:
            print('Opción inválida\n')
```
